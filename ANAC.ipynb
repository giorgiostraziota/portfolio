```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('prova.csv')
df.head()
```

```python
from bs4 import BeautifulSoup

# Applichiamo il parsing HTML a tutta la colonna "CIG"
def estrai_cig(html):
    soup = BeautifulSoup(html, "html.parser")
    return soup.get_text().strip()

# Applica la funzione a tutta la colonna
df["CIG"] = df["CIG"].apply(estrai_cig)

print(df[["CIG", "OGGETTO"]].head())

df.tail()
```

```python
df.info()
```

```python
def pulisci_html(valore):
    if pd.isnull(valore):  # gestisce i NaN
        return valore
    soup = BeautifulSoup(valore, "html.parser")
    return soup.get_text().strip()

# Colonne da ripulire
colonne_da_pulire = ["RUP", "AMMINISTRAZIONE APPALTANTE"]

# Applica la funzione a ciascuna colonna
for col in colonne_da_pulire:
    df[col] = df[col].apply(pulisci_html)

# Elimina le colonne non necessarie
df.drop(columns=["DATA AGGIUDICAZIONE", "VALORE AGGIUDICAZIONE"], inplace=True)

# Visualizza gli ultimi valori per conferma
df.tail()
```

```python
# Parole da cercare (anche all'interno di parole)
keywords = ["pane", "derrate", "alimenti", "zinco", "PANE"]

# Filtro: cerca se OGGETTO contiene una qualsiasi delle parole (anche come sottostringa)
mask = df["OGGETTO"].str.lower().apply(lambda testo: any(k in testo for k in keywords) if isinstance(testo, str) else False)

# Applichiamo il filtro
df_filtrato = df[mask]

# Mostra i risultati filtrati
df_filtrato
```
